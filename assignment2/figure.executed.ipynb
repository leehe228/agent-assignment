# PA2 Figure Notebook

This notebook creates report-ready figures from the values in `PA2_report.md`.

```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

fig_dir = Path("figures")
fig_dir.mkdir(parents=True, exist_ok=True)

summary_df = pd.DataFrame(
    {
        "Method": ["CoT", "ToT"],
        "SuccessRate": [1.0, 1.0],
        "SuccessCount": [10, 10],
        "Total": [10, 10],
    }
)
summary_df
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(summary_df["Method"], summary_df["SuccessRate"], color=["#2E86AB", "#F18F01"])

for bar in bars:
    h = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, h + 0.01, f"{h:.3f}", ha="center", va="bottom")

ax.set_ylim(0, 1.1)
ax.set_ylabel("Success Rate")
ax.set_title("Game of 24 Success Rate: CoT vs ToT")

out_path = fig_dir / "success_rate_comparison.png"
fig.savefig(out_path, dpi=200, bbox_inches="tight")
plt.show()
print(f"Saved: {out_path}")
```

```python
expr_compare_df = pd.DataFrame(
    {
        "Category": ["Same final expression", "Different final expression"],
        "Count": [6, 4],
    }
)
expr_compare_df
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(expr_compare_df["Category"], expr_compare_df["Count"], color=["#3FA34D", "#D1495B"])

for bar in bars:
    h = int(bar.get_height())
    ax.text(bar.get_x() + bar.get_width() / 2, h + 0.1, str(h), ha="center", va="bottom")

ax.set_ylim(0, max(expr_compare_df["Count"]) + 1)
ax.set_ylabel("Number of Instances")
ax.set_title("Expression Diversity Between CoT and ToT")
ax.tick_params(axis="x", rotation=15)

out_path = fig_dir / "expression_diversity.png"
fig.savefig(out_path, dpi=200, bbox_inches="tight")
plt.show()
print(f"Saved: {out_path}")
```
